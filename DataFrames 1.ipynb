## Broadcasting Operations

```python
nba = pd.read_csv("nba.csv") #reloads the original dataframe
```

```python
nba["Age"].add(5) #adds 5 to each value of age
                  #works also for NaN values
nba["Age"] + 5    #works exactly like the previous one

nba["Salary"].sub(10000) #substract 10000 from every salary value
nba["Salary"] - 10000    #works exactly like the previuos command

nba["Weight"].mul(0.453592)  #returns weight in kilograms
nba["Weight"] * 0.453592     #works exactly like the previuos command

nba["Weight in kilograms"] = nba["Weight"] * 0.453592 #creates a new column 
nba["Salary"].div(1000000)
nba["Salary in Millions"] = nba["Salary"] / 1000000 #creates a new column with salary in million
```

## Drop Rows with Null Values

```python
nba = pd.read_csv("nba.csv") #reloads the original dataframe   
```

```python
nba.dropna()  #removes any rows from the dataframe if there is one or more NaN value
              #if there is a row with any value of NaN it will be removed
```

```python
nba.dropna(how = "all") #drops rows if all values in a row are NaN                       
```

```python
nba.dropna(axis = 1) ####!!!!THIS SHOULD NOT REMOVE ALL THE COLUMNS BUT JUST THOSE WITH any NaN
```

```python
nba.dropna(subset = ["Salary", "College"]) #Removes rows where the salary and College valuea are NaN
```

## Fill in Null values with the .fillna() Method

```python
nba.fillna(0) #filling every NaN is not a good idea for different data types in a dataframe
              #it is better to use fillna on a Series
```

```python
nba["Salary"].fillna(0) #call fillna ona series for salary 
```

```python
nba["College"].fillna("No college")  #replace the NaN in college column with a string
```

```python
nba["Number"] = nba["Number"].astype("int") 
```

## Sort a DataFrame with the .sort_value() method, Part 1

```python
nba.sort_values("Name", ascending = False, inplace=True,na_position="last" )  #stores the entire dataframe by values in name series 
                                                        #descending order
                                                        #inplace
                                                        #NaN values will be placed at the end due to the na_position parameter
```

## Sort a DataFrame with the .sort_value() method, Multiple columns

nba = pd.read_csv("nba.csv")

```python
nba.head(3)
```

```python
nba.sort_values(["Team", "Name"], ascending = False)   #first sort by team and then sort by name
                                                       #in descending order
```

```python
nba.sort_values(["Team", "Name"], ascending = [True, False])  #team is sorted in alphabetical order
                                                              #name is sorted in descending order
```

## Rank Values with the .rank() Method

```python
nba = pd.read_csv("nba.csv").dropna(how="all")
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
```

```python
nba.head(4)
```

```python
nba["Salary rank"] = nba["Salary"].rank(ascending=False).astype("int")
nba.head(3)
```
